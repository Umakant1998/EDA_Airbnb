````python
# import pandas as pd
# import matplotlib.pyplot as plt
# import numpy as np

# # Load the Airbnb NYC 2019 dataset
# airbnb_df = pd.read_csv('AB_NYC_2019.csv')

# Split the dataset by room type
entire_home_apt = airbnb_df[airbnb_df['room_type'] == 'Entire home/apt']
private_room = airbnb_df[airbnb_df['room_type'] == 'Private room']
shared_room = airbnb_df[airbnb_df['room_type'] == 'Shared room']

# Create boxplots for price
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].boxplot(entire_home_apt['price'])
axs[0].set_title('Entire home/apt')
axs[1].boxplot(private_room['price'])
axs[1].set_title('Private room')
axs[2].boxplot(shared_room['price'])
axs[2].set_title('Shared room')
plt.show()

# Remove outliers from price for each dataset
def remove_outliers(df):
    Q1 = np.percentile(df['price'], 25)
    Q3 = np.percentile(df['price'], 75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
    return df

entire_home_apt = remove_outliers(entire_home_apt)
private_room = remove_outliers(private_room)
shared_room = remove_outliers(shared_room)

# Combine the datasets
combined = pd.concat([entire_home_apt, private_room, shared_room], axis=0)

# Show the summary statistics for each dataset
print('Entire home/apt:')
print(entire_home_apt['price'].describe())
print('\nPrivate room:')
print(private_room['price'].describe())
print('\nShared room:')
print(shared_room['price'].describe())
print('\nCombined:')
print(combined['price'].describe())
````

````python
# import pandas as pd
# import matplotlib.pyplot as plt
# import numpy as np

# # Load the Airbnb NYC 2019 dataset
# airbnb_df = pd.read_csv('AB_NYC_2019.csv')

# Split the dataset by room type
entire_home_apt = airbnb_df[airbnb_df['room_type'] == 'Entire home/apt']
private_room = airbnb_df[airbnb_df['room_type'] == 'Private room']
shared_room = airbnb_df[airbnb_df['room_type'] == 'Shared room']

# Create boxplots for price
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].boxplot(entire_home_apt['price'])
axs[0].set_title('Entire home/apt')
axs[1].boxplot(private_room['price'])
axs[1].set_title('Private room')
axs[2].boxplot(shared_room['price'])
axs[2].set_title('Shared room')
plt.show()

# Replace outliers with median for each dataset
def replace_outliers_with_median(df):
    Q1 = np.percentile(df['price'], 25)
    Q3 = np.percentile(df['price'], 75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    median = df['price'].median()
    df.loc[df['price'] > upper_bound, 'price'] = median
    df.loc[df['price'] < lower_bound, 'price'] = median
    return df

entire_home_apt = replace_outliers_with_median(entire_home_apt)
private_room = replace_outliers_with_median(private_room)
shared_room = replace_outliers_with_median(shared_room)

# Combine the datasets
combined = pd.concat([entire_home_apt, private_room, shared_room], axis=0)

# Show the summary statistics for each dataset
print('Entire home/apt:')
print(entire_home_apt['price'].describe())
print('\nPrivate room:')
print(private_room['price'].describe())
print('\nShared room:')
print(shared_room['price'].describe())
print('\nCombined:')
print(combined['price'].describe())
````
